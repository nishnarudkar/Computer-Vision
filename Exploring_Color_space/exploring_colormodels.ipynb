```python
import cv2
```

```python
bgr_image = cv2.imread("C:\\Users\\Asus\\OneDrive\\Desktop\\Computer_vision\\Exploring_Color_space\\butterfly.jpg")
if bgr_image is None:
    raise FileNotFoundError("Image not found: check path and filename")
print("Image shape:", bgr_image.shape)
```

```python
# Convert the BGR image to GrayScale image

gray_image= cv2.cvtColor(bgr_image,cv2.COLOR_BGR2GRAY)
cv2.imwrite("C:\\Users\\Asus\\OneDrive\\Desktop\\Computer_vision\\Exploring_Color_space\\gray_image_generated.jpg",gray_image)
```

```python
# Convert the BGR image to HSV image

hsv_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2HSV)
cv2.imwrite("C:\\Users\\Asus\\OneDrive\\Desktop\\Computer_vision\\Exploring_Color_space\\hsv_image_generated.jpg",hsv_image)
```

```python
# Convert the BGR image to LAB image

lab_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2LAB)
cv2.imwrite("C:\\Users\\Asus\\OneDrive\\Desktop\\Computer_vision\\Exploring_Color_space\\lab_image_generated.jpg",lab_image)
```

```python
# Convert the BGR image to RGB

rgb_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2RGB)
cv2.imwrite("C:\\Users\\Asus\\OneDrive\\Desktop\\Computer_vision\\Exploring_Color_space\\rgb_image_generated.jpg",rgb_image)
```
